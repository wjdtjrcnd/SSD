```python
import os
import numpy as np
import re

with open('positive_prior.txt') as f:
    x = f.read()

#    
flattened_data = re.findall(r'\d+', x)

#  
flattened_data = [int(num) for num in flattened_data]

print(flattened_data)
```

```python
import matplotlib.pyplot as plt

#   
plt.figure(figsize=(12, 8))  #  12,  8  

#  
plt.hist(flattened_data, bins=8732, color='skyblue', edgecolor='blue')

#     
plt.title('Histogram of all anchor boxes')

plt.xlabel('Value')
plt.ylabel('Frequency')

#  
plt.show()
```

```python
import matplotlib.pyplot as plt

#  
hist, bins, _ = plt.hist(flattened_data, bins=[0, 5775, 7941, 8541, 8691, 8727, 8731], color='skyblue', edgecolor='blue')

#    
bin_centers = 0.5 * (bins[:-1] + bins[1:])

#     
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

#    
for count, x in zip(hist, bin_centers):
    plt.text(x, count, str(int(count)), ha='center', va='bottom')

#  
plt.show()
```

```python
import matplotlib.pyplot as plt

# 
data = {
    '38*38': 3074,
    '19*19': 2804,
    '10*10': 1868,
    '5*5': 845,
    '3*3': 635,
    '1*1': 603
}

#  
bars = plt.bar(data.keys(), data.values(), color=['skyblue', 'orange', 'green', 'red', 'purple', 'pink'])

#    
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')

#     
plt.title('Feature Maps')
plt.xlabel('Feature Map')
plt.ylabel('Value')

#  
plt.show()
```

```python
prior_38, prior_19, prior_10, prior_5, prior_3, prior_1 = [], [], [], [], [], [] 
for i in range(len(flattened_data)):
    k=flattened_data[i]
    if 0<=k<=5775:
        prior_38.append(k)
    elif 5775<k<=7941:
        prior_19.append(k)
    elif 7941<k<=8541:
        prior_10.append(k)
    elif 8541<k<=8691:
        prior_5.append(k)
    elif 8691<k<=8727:
        prior_3.append(k)
    else:
        prior_1.append(k)
```
