```python
import os
import numpy as np

with open('best_prior_match.txt') as f:
    x = f.read()

z = x.split('\n')
y=[]
for i in range(len(z)):
    y.append(z[i][8:-2])
```

```python
# 문자열을 쉼표를 기준으로 나누고, 각 숫자들을 정수형으로 변환하여 리스트에 저장
y_1 = [list(map(int, item.split(','))) for item in y]

# 다차원 리스트를 1차원 리스트로 만들기
flattened_data = np.concatenate(y_1)
```

```python
import matplotlib.pyplot as plt

# 히스토그램 그리기
plt.hist(flattened_data, bins=8732, color='skyblue', edgecolor='blue')

# 그래프 제목과 축 라벨 설정
plt.title('Histogram of all anchor boxes')

plt.xlabel('Value')
plt.ylabel('Frequency')

# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt

# 히스토그램 그리기
hist, bins, _ = plt.hist(flattened_data, bins=[0, 5775, 7941, 8541, 8691, 8727, 8731], color='skyblue', edgecolor='blue')

# 각 구간의 중심값 계산
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# 그래프 제목과 축 라벨 설정
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 히스토그램 위에 값 표시
for count, x in zip(hist, bin_centers):
    plt.text(x, count, str(int(count)), ha='center', va='bottom')

# 그래프 출력
plt.show()
```

```python
prior_38, prior_19, prior_10, prior_5, prior_3, prior_1 = [], [], [], [], [], [] 
for i in range(len(flattened_data)):
    k=flattened_data[i]
    if 0<=k<=5775:
        prior_38.append(k)
    elif 5776<k<=7941:
        prior_19.append(k)
    elif 7942<k<=8541:
        prior_10.append(k)
    elif 8542<k<=8691:
        prior_5.append(k)
    elif 8692<k<=8727:
        prior_3.append(k)
    else:
        prior_1.append(k)
```

```python
a, b, c, d = 0, 0, 0, 0
for i in range(len(prior_38)):
    j=prior_38[i]
    if j%4==0:
        a+=1
    elif j%4==1:
        b+=1
    elif j%4==2:
        c+=1
    else:
        d+=1
print(a,b,c,d)
```
